__int()__: Constructor that takes as input a pair of Point objects that represent the ends points of the line segment

Length():: returns the length if the segment 

Slope() returns the slope of the segment of none if the slope is unbounded 

 

>>> p1 = Point(3,4)

>>> p2 = Point()

>>> s = Segment(p1,p2)

>>> s.length()

5.0

>>> s.slope()

0.75

```python
import math

class point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

class segment:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2

    def length(self):
        dx = self.p2.x - self.p1.x
        dy = self.p2.y - self.p1.y
        return math.sqrt(dx**2 + dy**2)

    def slope(self):
        dx = self.p2.x - self.p1.x
        dy = self.p2.y - self.p1.y

        if dx == 0:
            return None  
        return dx / dy

p1 = point(3, 4)
p2 = point()
s = segment(p1, p2)

print(s.length())  
print(s.slope()) 
```

+---------------+---------+
| Column Name   | Type    |
+---------------+---------+
| id            | int     |
| recordDate    | date    |
| temperature   | int     |
+---------------+---------+
id is the column with unique values for this table.
This table contains information about the temperature on a certain day.
 
 

Write a solution to find all dates' Id with higher temperatures compared to its previous dates (yesterday).

Return the result table in any order.

Please use the following input Weather table for your solution. To receive full marks you will need to create the database, create the table, insert the data below and execute the SQL query.

 

Input: 
Weather table:
+----+------------+-------------+
| id | recordDate | temperature |
+----+------------+-------------+
| 1  | 2015-01-01 | 10          |
| 2  | 2015-01-02 | 25          |
| 3  | 2015-01-03 | 20          |
| 4  | 2015-01-04 | 30          |
+----+------------+-------------+
Output: 
+----+
| id |
+----+
| 2  |
| 4  |
+----+
Explanation: 
In 2015-01-02, the temperature was higher than the previous day (10 -> 25).
In 2015-01-04, the temperature was higher than the previous day (20 -> 30).

```python
import sqlite3

conn = sqlite3.connect('mid_term.db')  

cursor = conn.cursor()

create_table_query = '''
CREATE TABLE Weather_table (
    id INT PRIMARY KEY,
    recordDate DATE,
    temperature INT
);
'''

insert_queries = [
    "INSERT INTO Weather_table VALUES (1, '2015-01-01', 10);",
    "INSERT INTO Weather_table VALUES (2, '2015-01-02', 25);",
    "INSERT INTO Weather_table VALUES (3, '2015-01-03', 20);",
    "INSERT INTO Weather_table VALUES (4, '2015-01-04', 30);"
]

select_query = '''
SELECT t1.id
FROM Weather_table t1
JOIN Weather_table t2 ON t1.id = t2.id + 1
WHERE t1.temperature > t2.temperature;
'''

# Execute the SQL queries
cursor.executescript(create_table_query)
for insert_query in insert_queries:
    cursor.execute(insert_query)

# Execute the SELECT query and fetch the results
cursor.execute(select_query)
results = cursor.fetchall()

# Print the results
for row in results:
    print(row)

# Commit the changes and close the database connection
conn.commit()
conn.close()
```

Develop a recursive function tough() that takes two nonnegative integer arguments and outputs a pattern as shown below. Hint: The first argument represents the indentation of the pattern, where the second argument -- always a pattern of 2 indicates the number *s in the longest line of *s in the pattern

```python
def tough(indentation, stars):
    if stars == 0:
        # Base case: If there are no stars, do nothing
        return
    if stars == 1:
        # Base case: Print a single '*' with the given indentation
        print(" " * indentation + "*")
    else:
        # Recursively print the upper part of the pattern
        tough(indentation, stars // 2)

        # Print the current line with the specified number of '*'s and indentation
        print(" " * indentation + "*" * stars)

        # Recursively print the lower part of the pattern
        tough(indentation + stars // 2, stars // 2)

# Example usage:
tough(0,0)
print("-----")
tough(0,1)
print("-----")
tough(0,2)
print("-----")
tough(0,4)
```
